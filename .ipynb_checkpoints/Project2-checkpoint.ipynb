# Part 1: Team ranking

```julia
using LinearAlgebra, DelimitedFiles
A = readdlm("games.csv", ',', Float32);
teams = readdlm("teams.csv", ',');

# This function takes in a vector containing ranking for each team, and
# output the names of the top k teams sorted by ranking
function sort_by_ranking(rankings; k=10)
    @assert length(rankings) == length(teams)
            "$(length(rankings)) rankings != $(length(teams)) teams"
    indices = sort(1:length(rankings), by = i -> -rankings[i])
    return collect(Iterators.take(zip(rankings[indices], teams[indices]), k))
end
```

```julia
### Your Code Here ###
function get_Q(A)
    m,n = size(A)
    Q = zeros(m,n) 
    for i in 1:m
        games_between = sum(A[i,:]) + sum(A[:,i])
        Q[i,:] = A[i,:]/games_between
    end
    return Q
end 
```

```julia
function get_r(Q)
    epsilon = 0.0000001
    m,n = size(Q)
    x = randn(m)
    y = normalize(Q*x)
    while norm(x-y) > epsilon
        x = y
        y = normalize(Q*x)
    end
    
    if all(>=(0), y)
        return y
    else
        return -y
    end
end
```

```julia
display(sort_by_ranking(get_r(get_Q(A)); k=20))
```

```julia
function getA(a, b)
    n = length(a)
    @assert a + b == ones(n) "Probabilities do not sum to 1!"
    A = zeros(Float64, (n,n))
    for i in 1:n
        if i == 1
            A[i,i] = a[i]
            A[i,i+1] = b[i]
            elseif i ==n 
            A[i,i-1] = a[i]
            A[i,i] = b[i]
                
        else 
            A[i,i-1] = a[i]
            A[i,i+1] = b[i]
        
        end 
    end 
    return transpose(A)
end
  a = ones(10)*.49
  b = ones(10)*.51

```

## 2c

```julia
# if you were to move "infinitely" what is the probability you'd be at a particular spot.
function stationary(a, b)
    ### Your Code Here ###
    A = getA(a,b)
    
    F = eigen(A)
    
    # find dominant eigen value, vector
    last = F.vectors[:,length(a)]
    
    # "normalize"
    p = last/sum(last)
    
    return p
    
end


a = zeros(200)
b = zeros(200)

for i in 1:200
    a[i] = 0.49
    b[i] = 0.51
end

display(plot(stationary(a,b), label = "Stationary distribution probabilities"))
```

```julia
function stationary_linear(a, b)
    W = getA(a,b)
    C = W-I
    return_value = nullspace(C)
    
    return_answer = return_value/sum(return_value)
    
    return return_answer
    
end   
    
display(nullspace(Matrix(1I,3,3)))
answer_to_plot = stationary_linear(a,b)

plot(answer_to_plot, label = "Stationary distribution probabilities")
```

## 2d

```julia
function generate_barrier(l, w; eps= 0.1)
    a = [ones(l)*1/2; ones(w)*(1/2-eps); ones(l)*1/2]
    b = [ones(l)*1/2; ones(w)*(1/2+eps); ones(l)*1/2]
    a,b
end

plot()
for i in 0:20
    a,b = generate_barrier(50, 50; eps = i/500)
    s = stationary(a,b)
    plot!(s)   
end
plot!(legend=false,ylim=(0,0.02))
```

## 2e

```julia
# This makes sense because epsilon has a bias towards movement toward the right, so if we are on the right already
# it makes sense that we converge to some sort of "equilibrim" faster.
function animtime(a, b, xinit, niters)
    x = xinit
    A = getA(a,b)
    fig = @gif for i in 1:niters
        x = A*x
        plot(x, ylim=(0,.05))
    end every 50
    fig
end

a,b = generate_barrier(50, 50; eps=0.1)
x_stationary  =  stationary(a, b)
len           =  size(x_stationary)[1]
x_leftmost    =  zeros(len)
x_rightmost   =  zeros(len)
x_leftmost[1] =  1
x_rightmost[len]= 1 
animtime(a, b, [x_stationary x_leftmost x_rightmost], 5000)
```

```julia
function animtimevecs(a, b, xinit, niters)
    x = xinit
    A = getA(a,b)
    vec = eigvecs(A,sortby=abs)
    len = length(A)
    Eigen = eigen(A)
    
    
    fig = @gif for i in 1:niters
        
        Eigen_matrix = diagm(Eigen.values)
       
        b = Eigen_matrix^i*vec^-1*x
        plot(abs.(b), ylim=(0,.2))
    end every 5
    fig
end

a, b = generate_barrier(20, 20; eps=0.1)
x = abs.(randn(length(a)))
animtimevecs(a, b, x/sum(x), 500)
```
